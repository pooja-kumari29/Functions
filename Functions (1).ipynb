1. What is the difference between a function and a method in Python?

   A function is a standalone block of code that is defined using the def keyword and can be called independently.
   A method is a function that is associated with an object (usually of a class) and is called using dot notation (.). It automatically takes the object (self) as its first argument.

```python
# example of function
def greet(name):
    return "Hello, " + name

print(greet("Pooja"))
```

```python
# example of method
class Person:
    def say_hello(self):
        return "Hello!"

p = Person()
print(p.say_hello())  # say_hello is a method
```

3. What are the different ways to define and call a function in Python?

   In Python, functions are reusable blocks of code used to perform specific tasks. There are several ways to define and call functions based on different requirements.

  - Basic Function Definition and Call -
   A simple function is defined using the def keyword and called by its name followed by parentheses.
   
  - Function with Parameters -
   Functions can take input values, called parameters.
   
  - Function with Return Statement -
   A function can return a value using the return keyword.

   - Function with Default Parameters -
  If no argument is passed, default values are used.

   - Function with Keyword Arguments -
  Keyword arguments allow you to pass values by parameter name.

   - Variable-Length Arguments -
  These allow functions to accept any number of arguments.

  - Positional Arguments using *args
  - Keyword Arguments using **kwargs

   - Lambda (Anonymous) Function -
  Used for small, one-line functions.

  - Function Calling Another Function -
  One function can call another inside it.


```python
# 1. Basic Function Definition and Call
def greet():
    print("Hello!")

greet()  # Calling the function


# 2. Function with Parameter
def greet(name):
    print("Hello,", name)

greet("Pooja")  # Output: Hello, Pooja

# 3. Function with Return Statement
def add(a, b):
    return a + b

result = add(5, 3)
print(result)  # Output: 8

# 4. Function with Default Parameters
def greet(name="Guest"):
    print("Hello,", name)

greet()           # Output: Hello, Guest
greet("Pooja")    # Output: Hello, Pooja

# 5. Function with Keyword Arguments
def profile(name, age):
    print(name, "is", age, "years old")

profile(age=25, name="Pooja")  # Output: Pooja is 25 years old

# 6. Positional Arguments using *args
def show_numbers(*args):
    for num in args:
        print(num)

show_numbers(1, 2, 3)  # Output: 1 2 3
# 6. Keyword Arguments using **kwargs
def show_details(**kwargs):
    for key, value in kwargs.items():
        print(key, ":", value)

show_details(name="Pooja", age=25)

# 7. Lambda (Anonymous) Function

square = lambda x: x * x
print(square(4))  # Output: 16

# 8. Function Calling Another Function
def square(x):
    return x * x

def display():
    result = square(5)
    print("Square is:", result)

display()  # Output: Square is: 25
```

4. What is the purpose of the `return` statement in a Python function?

   The return statement in a Python function serves two main purposes:

  - It ends the function execution: When a return statement is encountered, the function stops running, and control is passed back to the place where the function was called.

  - It sends a value back to the caller: You can use return to pass data or a result from the function to the calling code.

  In this example:

  - The function add calculates the sum of a and b.

  - The return statement sends that sum back to the caller.

  - The value 8 is stored in the variable result.

```python
def add(a, b):
    return a + b

result = add(3, 5)
print(result)  # Output: 8
# If no return statement is used, the function returns None by default.
```

5.  What are iterators in Python and how do they differ from iterables?

   In Python, iterables and iterators are related concepts used in looping and iteration, but they are not the same.

    Iterable:
  An iterable is any object that can return its elements one at a time. Examples: list, tuple, string, set, dictionary, etc.

  - It implements the __iter__() method.

  - It can be passed to functions like for, in, list(), etc.

    Iterator:
  An iterator is an object that represents a stream of data and returns the next value with each call to next().

  - It implements both __iter__() and __next__() methods.

  - It remembers its state during iteration.

```python
my_list = [1, 2, 3]  # This is an iterable
my_list = [1, 2, 3]
it = iter(my_list)     # 'it' is now an iterator

print(next(it))  # Output: 1
print(next(it))  # Output: 2
print(next(it))  # Output: 3
```

6.  Explain the concept of generators in Python and how they are defined.

  A generator in Python is a special type of iterator that is used to generate values on the fly (one at a time) using the yield keyword, rather than storing all values in memory like lists.

  - Generators produce values lazily, meaning they generate the next value only when requested.

  - They are memory efficient and suitable for large datasets or infinite sequences.

  - Generators automatically implement the iterator protocol (__iter__() and __next__()).

   Generators are defined in two ways:

  - Using a Generator Function (with yield) - The yield statement pauses the function and remembers its state until the next value is requested.

  - Using a Generator Expression-
  Similar to list comprehensions but with parentheses.

```python
def count_up_to(n):
    count = 1
    while count <= n:
        yield count
        count += 1

# Create generator
gen = count_up_to(3)

# Get values
print(next(gen))  # Output: 1
print(next(gen))  # Output: 2
print(next(gen))  # Output: 3
# next(gen) now would raise StopIteration

# using parenthesis
gen_exp = (x * x for x in range(5))
print(next(gen_exp))  # Output: 0
print(next(gen_exp))  # Output: 1
```

7.  What are the advantages of using generators over regular functions?

  Advantages of Using Generators Over Regular Functions in Python:
    
  - Memory Efficiency- Generators do not store the entire sequence in memory.
They generate values on the fly using yield, which is ideal for large datasets or infinite sequences.

  - Lazy Evaluation -
Values are computed only when needed, not all at once.
This reduces computation and memory load when not all results are required.

  - Improved Performance -
Generators are generally faster than functions returning large lists, especially when dealing with loops and large datasets.

  - Clean and Simple Code -
Using yield makes complex iterative logic easier to read and manage compared to building lists manually.

  - Infinite Sequences Support -
You can easily define sequences that never end without running out of memory.



8. What is a lambda function in Python and when is it typically used?

  A lambda function in Python is a small anonymous function defined using the lambda keyword instead of the standard def keyword. It is typically used for short, throwaway functions that are not reused elsewhere.

  Typical Use Cases:
  
  Lambda functions are commonly used when:

  - You need a simple function for a short period.

  - You are using functions like map(), filter(), or sorted() where a function is passed as an argument.

  Examples:
  - With map()

  - With filter()

  - With sorted() (sorting by the second item in a tuple)

```python
# with map()
nums = [1, 2, 3, 4]
squares = list(map(lambda x: x**2, nums))
print(squares)  # Output: [1, 4, 9, 16]

# with filter()
nums = [1, 2, 3, 4]
squares = list(map(lambda x: x**2, nums))
print(squares)  # Output: [1, 4, 9, 16]

# with sorted()
pairs = [(1, 3), (2, 2), (3, 1)]
sorted_pairs = sorted(pairs, key=lambda x: x[1])
print(sorted_pairs)  # Output: [(3, 1), (2, 2), (1, 3)]
```

9. Explain the purpose and usage of the `map()` function in Python

  The map() function is used to apply a given function to all items in an iterable (like a list, tuple, etc.) and return a new map object (which is an iterator).

  - function: A function (can be built-in, user-defined, or lambda) that will be applied to each item.

  - iterable: A sequence like a list, tuple, etc.
  
  It processes each element of the iterable with the given function and returns a map object, which can be converted to a list, tuple, etc.


```python
# Using map() with a built-in function
numbers = [1, 2, 3, 4]
result = map(str, numbers)  # Converts all numbers to strings
print(list(result))  # Output: ['1', '2', '3', '4']

# Using map() with a lambda function
numbers = [1, 2, 3, 4]
squares = map(lambda x: x**2, numbers)
print(list(squares))  # Output: [1, 4, 9, 16]

# Using map() with multiple iterables
a = [1, 2, 3]
b = [4, 5, 6]
result = map(lambda x, y: x + y, a, b)
print(list(result))  # Output: [5, 7, 9]
```

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

  n Python, the map(), filter(), and reduce() functions are used for functional programming, where operations are applied to sequences like lists or tuples.

  - map(function, iterable) -
  Purpose: Applies a function to each item in an iterable and returns a new iterable (map object).

  Use case: When you want to transform or modify all items.

  - filter(function, iterable) -
  Purpose: Filters elements from an iterable for which the function returns True.

  Use case: When you want to select specific items based on a condition.

  - reduce(function, iterable) -
  Purpose: Applies a function cumulatively to the items in an iterable, reducing it to a single value.

  Use case: When you want to combine all items into one result (like sum, product, etc.).

```python
# 4. Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(n):
    if n <= 1:
        return False  # 0 and 1 are not prime
    for i in range(2, int(n**0.5) + 1):  # Check up to square root of n
        if n % i == 0:
            return False
    return True

# Example usage:
for num in range(1, 10):
    if is_prime(num):
        print(f"{num} is a prime number.")
    else:
        print(f"{num} is not a prime number.")
```

```python
# 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
class FibonacciIterator:
    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.n1, self.n2 = 0, 1  # Starting values
        self.count = 0

    def __iter__(self):
        return self  # The iterator object itself

    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration  # End of iteration

        if self.count == 0:
            self.count += 1
            return self.n1
        elif self.count == 1:
            self.count += 1
            return self.n2
        else:
            next_val = self.n1 + self.n2
            self.n1, self.n2 = self.n2, next_val
            self.count += 1
            return next_val

# Example usage:
fib = FibonacciIterator(10)
for num in fib:
    print(num)
```

```python
# 6.  Write a generator function in Python that yields the powers of 2 up to a given exponent.
def power_of_two(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

# Example usage:
for power in power_of_two(5):
    print(power)
```

```python
# 10. Create a Python program that uses `filter()` to remove all the vowels from a given string.
def remove_vowels(input_string):
    vowels = 'aeiouAEIOU'
    # Filter out characters that are vowels
    no_vowels = filter(lambda char: char not in vowels, input_string)
    # Join the remaining characters back into a string
    return ''.join(no_vowels)

# Example usage
text = "Hello, how are you?"
result = remove_vowels(text)
print("String without vowels:", result)
```
